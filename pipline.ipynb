```python
from detectron2.utils.logger import setup_logger
setup_logger()


from matplotlib import pyplot as plt
from PIL import Image, ImageDraw
import numpy as np
import os, json, cv2
import math, time


import predict_module as pm 
```

```python
predictor = pm.predict_model(0.7)
```

```python
capture = cv2.VideoCapture('my_video.mp4')
fps = capture.get(cv2.CAP_PROP_FPS)		# 影片的幀率FPS
total_frame = capture.get(cv2.CAP_PROP_FRAME_COUNT)	# 影片的總幀數
print(fps)
print(total_frame)
```

```python
for i in range(0,int(total_frame)):
    success = capture.grab()
    if (i % 21)==0:
        print(i)
        ret, frame = capture.retrieve()
        cv2.imshow('frame',frame)
        cv2.waitKey(1)
        start = time.time()
        im2 = frame.copy()
        outputs_2 = predictor(im2)
        new_predict = pm.tidy_data(outputs_2)
        my_dict,image = pm.rebuild_bbox(im2,new_predict)


        pred_masks = outputs_2["instances"].pred_masks.numpy().copy()
        print(my_dict)
        print('\n')
        shrimp_length_width = pm.caculate_width_length_area(my_dict,pred_masks)
        print(shrimp_length_width)
        print('\n')
        df = pm.statistic_shrimp(shrimp_length_width)
        result = df.to_json(orient="index")
        parsed = json.loads(result)
        print(parsed)
        cv2.imshow('frame',image)
        cv2.waitKey(3)
        end = time.time()
        
        print(f"Total spend {end-start}")
        print('\n')
        print('\n')
        time.sleep(1.5)


cv2.destroyAllWindows()

```
